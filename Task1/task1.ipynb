```python
import nltk
from nltk.corpus import wordnet
nltk.download("wordnet")

lemmatizer = nltk.stem.wordnet.WordNetLemmatizer()
lemmatizer
```

```python
import requests

url = DATA_URL
r = requests.get(url, headers={'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8'}, timeout=15)
raw_text = r.text
raw_text
```

```python
import re

clean_pattern = re.compile("End of the Project Gutenberg EBook.*")
cleaner_text =  re.sub(clean_pattern, "", raw_text.replace("\n", " ").replace("\r", " "))
cleaner_text[-100:]
```

```python
from nltk.tokenize import word_tokenize
from nltk.tokenize import PunktSentenceTokenizer
import nltk
nltk.download('punkt')

tokens =  word_tokenize(cleaner_text)
"A total of %d 'tokens'" % len(tokens)
```

Task 1

```python
from tqdm.notebook import tqdm

lemmas = [lemmatizer.lemmatize(lemma) for lemma in tqdm(tokens) if str.isalpha(lemma)]
lemmas[-10:]
```

```python
len(lemmas)
```

Task 2

```python
import nltk
from nltk import FreqDist
from nltk.corpus import stopwords
nltk.download("stopwords")
STOPWORDS = set(stopwords.words("english"))
stopwords.words("english")[:5]
```

```python
fdist = FreqDist(lemmas)
top_100 = fdist.most_common(50)
top_100
```

```python
counter = 0.0
for token in top_100:
  if token[0] in STOPWORDS:
    counter += 1

counter/50
```
